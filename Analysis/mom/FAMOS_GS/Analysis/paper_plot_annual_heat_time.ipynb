```python
%matplotlib notebook
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly as py
import plotly.graph_objs as go
import ipywidgets as widgets
import seaborn as sns
import colorlover as cl

py.offline.init_notebook_mode(connected=True)
```

```python
fname0 = '../../../../../Analysis/mom/FAMOS_GS/Analysis/data/ITU-MOM/ITU-MOM_GSexps_1980_2009.nc'
ds0 = xr.open_mfdataset(fname0)

fname1 = '../../../../../Analysis/mom/FAMOS_GS/Analysis/data/MITB500/MIT_MITB500_1980_2009.nc'
ds1 = xr.open_mfdataset(fname1)

fname2 = '../../../../../Analysis/mom/FAMOS_GS/Analysis/data/Brest/Brest_GSexps_1980_2009.nc'
ds2 = xr.open_mfdataset(fname2)

fname3 = '../../../../../Analysis/mom/FAMOS_GS/Analysis/data/NorESM/NorESM_GS_experiments_1980_2009.nc'
ds3 = xr.open_mfdataset(fname3)

fname4 = '../../../../../Analysis/mom/FAMOS_GS/Analysis/data/FESOM/FESOM_GS_exps_1980_2009.nc'
ds4 = xr.open_mfdataset(fname4)
```

```python
dates0 = pd.date_range('1980-01', freq='M', periods=12*29)
ds0['time'] = dates0
ds1['time'] = dates0
ds2['time'] = dates0
ds3['time'] = dates0
ds4['time'] = dates0
```

```python
""" Plot annual mean of total heat fluxes anomaly"""
modelnames = ['ITU' , 'MITB500', 'Brest', 'NorESM', 'FESOM']
#colors = ['blue', 'darkorange', 'red', 'green', 'cyan']
colors = cl.scales['5']['qual']['Dark2']; 

def plot_bso_heat_total(models):
    
    datap = []
    datan = []
    layout = go.Layout(
    title='<b>Annual total heat transport anomaly @BSO</b>',
        yaxis=dict(
            title='Heat transport [TW]',
            #range=[-14, 14]
        ),
        xaxis=dict(
            title='year'
        )
    )  
    # names of the control, gsp and gsn variables [total]
    ctln = ['ctlbarents_heattotal', 'barht_c[0]', 'BSOrefheat', 'ht_NorESM_BSO_ctrl', 'ctlheattransp_BSO']
    gspn = ['gspbarents_heattotal', 'barht_gsp[0]', 'BSOgspheat', 'ht_NorESM_BSO_gsp', 'gspheattransp_BSO']
    gsnn = ['gsnbarents_heattotal', 'barht_gsm[0]', 'BSOgsnheat', 'ht_NorESM_BSO_gsn', 'gsnheattransp_BSO']

    coeffs = [1, 1, 1, 1, 1]
    for ind,mdl in enumerate(models):
        dnm = 'ds'+np.str(ind)+'.'+ctln[ind]+'.groupby(\'time.year\').mean().year'
        time = np.copy(eval(dnm))
        dnm = 'ds'+np.str(ind)+'.'+ctln[ind]+'.groupby(\'time.year\').mean()'
        dnm2 = 'ds'+np.str(ind)+'.'+gspn[ind]+'.groupby(\'time.year\').mean()'
        dnm3 = 'ds'+np.str(ind)+'.'+gsnn[ind]+'.groupby(\'time.year\').mean()'
        val1 = (np.copy(eval(dnm2)-eval(dnm))*coeffs[ind]) #/(np.copy(eval(dnm))*coeffs[ind])
        val2 = (np.copy(eval(dnm3)-eval(dnm))*coeffs[ind]) #/(np.copy(eval(dnm))*coeffs[ind])
        trace1 = go.Scatter(            
            x = time,
            y = val1,
            mode = 'lines',
            name = mdl+'-GSP',
            line = dict(
                width = 2.5,
                color = colors[ind],
                shape = 'smooth'
            )
        )
        datap.append(trace1)
        trace2 = go.Scatter(            
            x = time,
            y = val2,
            mode = 'lines',
            name = mdl+'-GSM',
            line = dict(
                width = 2.5,
                color = colors[ind],
                dash = 'dash',
                shape = 'smooth'
            )
        )
        datan.append(trace2)


   
    data = datap + datan
    
    fig = go.Figure(data=data, layout=layout)
    py.offline.iplot(fig)
    

def plot_fram_heat_total(models):
    
    datap = []
    datan = []
    layout = go.Layout(
    title='<b>Annual total heat transport anomaly @Fram Strait</b>',
        yaxis=dict(
            title='Heat transport [TW]',
            #range=[-14, 14]
        ),
        xaxis=dict(
            title='year'
        )
    )  
    # names of the control, gsp and gsn variables [total]
    ctln = ['ctlfram_heattotal', 'framht_c[0]', 'Framrefheat', 'ht_NorESM_FS_ctrl', 'ctlheattransp_FS']
    gspn = ['gspfram_heattotal', 'framht_gsp[0]', 'Framgspheat', 'ht_NorESM_FS_gsp', 'gspheattransp_FS']
    gsnn = ['gsnfram_heattotal', 'framht_gsm[0]', 'Framgsnheat', 'ht_NorESM_FS_gsn', 'gsnheattransp_FS']

    coeffs = [1, 1, 1, 1, 1]
    for ind,mdl in enumerate(models):
        dnm = 'ds'+np.str(ind)+'.'+ctln[ind]+'.groupby(\'time.year\').mean().year'
        time = np.copy(eval(dnm))
        dnm = 'ds'+np.str(ind)+'.'+ctln[ind]+'.groupby(\'time.year\').mean()'
        dnm2 = 'ds'+np.str(ind)+'.'+gspn[ind]+'.groupby(\'time.year\').mean()'
        dnm3 = 'ds'+np.str(ind)+'.'+gsnn[ind]+'.groupby(\'time.year\').mean()'
        val1 = (np.copy(eval(dnm2)-eval(dnm))*coeffs[ind]) #/(np.copy(eval(dnm))*coeffs[ind])
        val2 = (np.copy(eval(dnm3)-eval(dnm))*coeffs[ind]) #/(np.copy(eval(dnm))*coeffs[ind])
        trace1 = go.Scatter(            
            x = time,
            y = val1,
            mode = 'lines',
            name = mdl+'-GSP',
            line = dict(
                width = 2.5,
                color = colors[ind],
                shape = 'smooth'
            )
        )
        datap.append(trace1)
        trace2 = go.Scatter(            
            x = time,
            y = val2,
            mode = 'lines',
            name = mdl+'-GSM',
            line = dict(
                width = 2.5,
                color = colors[ind],
                dash = 'dash',
                shape = 'smooth'
            )
        )
        datan.append(trace2)

   
    data = datap + datan    
    fig = go.Figure(data=data, layout=layout)
    py.offline.iplot(fig)
    


plot_bso_heat_total(modelnames)
plot_fram_heat_total(modelnames)


#models = widgets.SelectMultiple(options=list(modelnames), value=modelnames, description='Models')    
#widgets.interactive(update_plot,models=models)
```
